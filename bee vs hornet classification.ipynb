```python
import tensorflow as tf
import numpy as np
import cv2
import os
import random
import matplotlib.pyplot as plt
```

```python
DIRECTORY= r"D:\8th semester\Machine Learning Lab\dataset\kaggle_bee_vs_wasp"
CATAGORIES= ['bees','hornets']
```

```python
data=[]  #list
for categories in CATAGORIES:
    folder=os.path.join(DIRECTORY,categories)
    label=CATAGORIES.index(categories)
    #print(folder)
    for img in os.listdir(folder):
        img=os.path.join(folder,img)
        #print(img)
        img_arr=cv2.imread(img)
        img_arr=cv2.resize(img_arr,(100,100))
        
        data.append([img_arr,label])
```

```python
data
```

```python
random.shuffle(data)
```

```python
x=[]
y=[]


for features,label in data:
    x.append(features)
    y.append(label)

X= np.array(x)
Y=np.array(y)
x
```

```python
X=X/255
X
```

```python
X.shape
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Activation
```

```python
model=Sequential()
model.add( Conv2D(64,(3,3),input_shape=X.shape[1:],activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

# model.add( Conv2D(32,(3,3),activation='relu'))
# model.add(MaxPooling2D(pool_size=(2,2)))


model.add(Flatten())

model.add(Dense(2,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=20,validation_split=0.30)
```

```python
model.summary()
```

```python
from keras.preprocessing import image
import keras.utils as image
import numpy as np
img_pred=image.load_img(r"D:\8th semester\Machine Learning Lab\dataset\kaggle_bee_vs_wasp\hornet\256220148_c5725ced57_m.jpg",target_size=(100,100))

img_pred=image.img_to_array(img_pred)
img_pred=np.expand_dims(img_pred, axis=0)


rslt= model.predict(img_pred)

print(rslt)
if rslt[0][0]>rslt[0][1]:
    prediction="hornet"
    
    
else:
    prediction="bee"
print(prediction)
```
